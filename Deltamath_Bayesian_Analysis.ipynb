##Commentary: Overall Probability of Success Quickly Rises After a User's First Attempt At a Problem.

##Next Step: Bayes
###Restrict to Users Who Got Question 1 Correct; Now what are subsequent Probabilities?

```python
#Methodology:
#1. Find User ID's Where Attempt=1 and Correct=1
#2. Rescan dataframe looking for these User ID's
#3. Groupby Attempt # and recalculate means

successful_students = all_attempts[(all_attempts.attempt_number==1) & (all_attempts.Correct==1)]['Student_ID']
attempts_with_initial_success = all_attempts[all_attempts.Student_ID.isin(successful_students)]
attempts_with_initial_success.groupby('attempt_number')['Correct'].mean()[:20] - all_attempts.groupby('attempt_number')['Correct'].mean()[:20]
```

###Remarks: Chance of Success on 2nd Attempt is ~15% higher for those with 1st success than versus the entire population

```python
##What about success on 1st and 2nd attempts?
##Ultimately we want to estimate our confidence that a student has "mastered" a skill.

success1_students = all_attempts[(all_attempts.attempt_number==1) & (all_attempts.Correct==1)]['Student_ID']
success2_students = all_attempts[(all_attempts.attempt_number==2) & (all_attempts.Correct==1)]['Student_ID']
success3_students = all_attempts[(all_attempts.attempt_number==3) & (all_attempts.Correct==1)]['Student_ID']
success4_students = all_attempts[(all_attempts.attempt_number==4) & (all_attempts.Correct==1)]['Student_ID']
success5_students = all_attempts[(all_attempts.attempt_number==5) & (all_attempts.Correct==1)]['Student_ID']
attempts_with_initial_success = all_attempts[(all_attempts.Student_ID.isin(success1_students)) & (all_attempts.Student_ID.isin(success2_students)) & (all_attempts.Student_ID.isin(success3_students)) & (all_attempts.Student_ID.isin(success4_students)) & (all_attempts.Student_ID.isin(success5_students))]
attempts_with_initial_success.groupby('attempt_number')['Correct'].mean()[:20]
```

```python
print "Overall Average:", attempts_with_initial_success['Correct'].mean()
print "Average on First 5 Attempts:", attempts_with_initial_success[attempts_with_initial_success.attempt_number<6]['Correct'].mean()
print "Average after First 5 Attempts:", attempts_with_initial_success[attempts_with_initial_success.attempt_number>5]['Correct'].mean()
print '\n'
print "Average for all Users after 5 Attempts", all_attempts[all_attempts.attempt_number>5]['Correct'].mean()
```
